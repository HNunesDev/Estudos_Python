```python
from tensorflow import keras
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris(return_X_y = True)
x=iris[0]
y=iris[1]
x
```

```python
datasets.load_iris()['feature_names']
```

```python
datasets.load_iris()['target_names']
```

```python
#Plotando o y de acordo com o tamanho das petalas
sns.scatterplot(x=x[:,2], y=x[:,3], hue=y, palette='tab10')
plt.xlabel('Comprimento (cm)', fontsize=16)
plt.ylabel('Largura (cm)', fontsize=16)
plt.title('Distribuiçao pétalas', fontsize=18)
plt.show()
```

```python
#plotando o y de acordo com as sépalas agora
sns.scatterplot(x=x[:,0], y=x[:,1], hue=y, palette='tab10')
plt.xlabel('Comprimento (cm)', fontsize=16)
plt.ylabel('Largura (cm)', fontsize=16)
plt.title('Distribuiçao sépalas', fontsize=18)
plt.show()
```

```python
#Categorizando o valor de Y
y = keras.utils.to_categorical(y)
y
```

```python
#Normalizando entrada
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
#Separando o conjunto 80/20
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.2, stratify = y, random_state=42)
```

```python
modelo = keras.Sequential([keras.layers.InputLayer(input_shape=[4,], name='Entrada'),
                           keras.layers.Dense(512, activation='relu', name='Oculta',
                                              kernel_initializer=keras.initializers.RandomNormal(seed=142)),
                            keras.layers.Dense(3, activation='softmax', name='Saida')
                           ])
```

```python
modelo.summary()
#O resultado de parametro é calculado:
#n de parametro vezes a quantidade de parametro mais o n de bias
#neste caso: (4x512) + 512
```

```python
#Compilando o modelo
#Etapa para definir a funcao de perda e a funcao de otimizacao
modelo.compile(loss = 'categorical_crossentropy',
               optimizer = 'rmsprop', #Este já é o otimizador padrao, mas por boa pratica declarei
               metrics = ['categorical_accuracy']) 
```

```python
#Treinando
historico = modelo.fit(x_treino, y_treino, epochs=100, validation_split=0.3)
```

```python
#Avaliando o aprendizado
historico.history
```

```python
pd.DataFrame(historico.history).plot()
plt.grid()
plt.show()
```

```python
#Utilizando o evaluate
modelo.evaluate(x_teste,y_teste)
```

```python
#Realizando predicao
entrada = np.array([[0.61, 0.5 ,0.69 ,0.79]])
modelo.predict(entrada) #Deve retornar uma Virginica, 3 opcao
```
